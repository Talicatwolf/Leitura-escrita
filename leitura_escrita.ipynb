```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("videos_analysis").getOrCreate()
```

```python
spark_df = spark.createDataFrame(df_parquet)
spark_df.createOrReplaceTempView("tb_videos")
```

```python
display(spark.catalog.listTables())
```

```python
result_df = spark.sql("SELECT * FROM tb_videos")
display(result_df)
```
